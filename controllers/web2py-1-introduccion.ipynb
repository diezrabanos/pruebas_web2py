## Aplicaciones
Una aplicación Web2py está compuesta por:
- Modelos: ficheros ".py" que contienen una descripción de la representación de los datos. Los datos que vemos
- Vistas: ficheros ".html" que contienen una descripción de la presentación de los datos. Cómo vemos  los datos
- Controladores: ficheros ".py" que contienen la lógica de negocio y el flujo de trabajo. Pueden contener cualquier objeto Python, pero unos en concreto son muy importantes, las acciones
  - Acciones: Funciones sin parámetros, dentro de los controladores, que reciben las peticiones de los usuarios y generan la respuesta oportuna
- Módulos: colecciones de objetos reutilizables (*.py)
- Ficheros estáticos: imágenes, scripts, hojas de estilo, etc.
- ...


### Resumen del flujo de la aplicación
- El usuario solicita una URL en su navegador, por ejemplo: http://servidor/aplicacion/controlador/accion
- La petición le llega al **servidor**, que se la pasa a Web2py
- Web2py analiza la petición y se la deriva a la **aplicación** oportuna. Puede haber varias aplicaciones en el mismo servidor
- Se ejecutan los **modelos** de la aplicación, lo que genera una representación de los datos con los que se trabajará, antes de pasar la petición al **controlador**
- Los controladores contienen funciones-acciones asociadas a una URL que son llamadas cuando se visita dicha URL
- La petición del usuario y el modelo de datos le llegarán al controlador oportuno, que ejecutará su código global y la **acción** solicitada por el usuario
- Si una acción devuelve algo distinto a una cadena de texto, Web2py tratará de renderizar el contenido devuelto con la **vista** asociada, que por defecto coincidirá en nombre con la acción, lo que genera una respuesta
- La respuesta es enviada de vuelta al usuario

## URLs Web2py
## host[:puerto]/aplicacion/[controlador]/[accion[.formato]][argumentos][?variables]
## Argumentos: /argumento1/argumento2/.../argumenton
## Variables: ?variable1=valor1&...&variablen=valorn
Por convenio:
  - aplicación por defecto: init o welcome
  - controlador por defecto: default
  - acción por defecto: index
  - extensión por defecto: html

**Ejemplos**:
- http://host -> http://host/init/default/index.html o http://host/welcome/default/index.html
- http://host/inventario/ -> http://host/inventario/default/index.html
- http://host/inventario/usuario -> http://host/inventario/usuario/index.html
- http://host/inventario/usuario/login -> http://host/inventario/usuario/login.html
- http://host/inventario/exporta/equipos.json -> http://host/inventario/exporta/equipos.json
- http://host/inventario/equipos/lista/nombre
- http://host/inventario/equipos/lista.json/nombre
- http://host/inventario/equipos/edita?id=123456

## Nuestra primera aplicación (app0)
La forma más rápida de crear una aplicación consiste en copiar la aplicación "welcome". Para ello basta con copiar el directorio que la contiene. Todas las aplicaciones están en el directorio "applications".

"welcome" -> "app0"

El directorio "app0" ahora contiene la aplicación "app0", que es un clon de "welcome". Está operativa sin necesidad de reiniciar el servidor.

Cada aplicación contendrá una serie de directorios y ficheros que la definen. En nuestro caso vamos a modificar su funcionamiento tocando los ficheros siguientes:
- controllers/default/index -> Hola mundo
```python
def index():
    '''
    app0 -> "Hola mundo"
    app/default -> "Hola mundo"
    app0/default/index -> "Hola mundo"
    '''

    # esta acción devuelve una cadena de texto que se envía directamente al cliente
    return 'Hola mundo'
```
- controllers/default/peticion -> Análisis de la petición (request)
```python
def peticion():
    'app0/default/peticion -> request'

    # la petición nos llega mediante el objeto request
    # r = request  # no es una cadena de texto
    r = str(request)  # ahora ya es una cadena de texto
    r = BEAUTIFY(request)  # objeto -> HTML
    r = HTML(BODY(BEAUTIFY(request)))  # HTML

    return r
```
- controllers/default/entorno -> Análisis del entorno (request.env)
```python
def entorno():
    'app0/default/entorno -> request.env'

    return HTML(BODY(BEAUTIFY(request.env)))
```
- controllers/default/args_vars -> Análisis de los argumentos y variables
```python
def args_vars():
    'app0/default/args_vars -> request.args + request.vars'

    return HTML(BODY(H2('Argumentos'),
                     BEAUTIFY(request.args),
                     HR(),
                     H2('Variables'),
                     BEAUTIFY(request.vars)))
```
- controllers/default/traduce -> i18n
```python
def traduce():
    'app0/default/traduce/palabra -> palabra_traducida'

    palabra = request.args(0) or ''
    palabra_traducida = T(palabra)

    return 'Palabra: ' + palabra + '. Traducción: ' + palabra_traducida
```
- controllers/default/sesion -> contador de visitas
```python
def sesion():
    'app0/default/sesion -> sesion'

    # el objeto session nos permite gestionar variables de sesión
    # podemos crear tantas variables como necesitemos
    if session.contador:
        # si ya existe la variable de sesión "contador" le añadimos 1
        session.contador = session.contador + 1
    else:
        # si no existe la varible de sesión "contador" la creamos con valor 1
        session.contador = 1

    # comprueba su funcionamiento visitando la URL de forma anómima
    # o desde varios navegadores
    return 'Contador: ' + str(session.contador)
```
- controllers/default/google -> redirección a Google
```python
def google():
    'app0/default/google -> www.google.com'

    # redirect(url) redirige a url
    redirect('http://www.google.com')
```
- controllers/default/traduce2 -> redirección interna
```python
def traduce2():
    'app0/default/traduce2 -> traduce/welcome'

    redirect(URL('traduce', args='welcome'))
```
- controllers/default/vistas -> vistas Web2py
```python
def vistas():
    'app0/default/vistas -> renderizado'

    x = 1
    y = 2

    # como la acción devuelve un diccionario
    # Web2py buscará la vista oportuna para renderizarlo
    # por convenio será app0/views/default/vistas.html
    # todas las vistas del controlador default de la aplicación app0
    # se encontrarán en applications/app0/views/default
    # si la vista no existe, el sistema tiende a utilizar una vista genérica
    # las vistas genéricas están en app0/views y se llamarán generic.*
    # así generic.html será la vista genérica para el caso de vistas html
    # locals() -> devuelve un diccionario con las variables locales a la función
    # en este caso locals() == {'x': 1, 'y': 2} == dict(x=1, y=2)
    return locals()
```
- views/default/vistas.html -> vista para controllers/default/vistas.html
```html
<!DOCTYPE html>
<html>
<head>
	<meta charset="utf-8">
	<meta http-equiv="X-UA-Compatible" content="IE=edge">
	<title>Vistas con Web2py</title>
</head>
<body>
<h3>Parámetros recibidos por la vista</h3>
x = {{=x}}
<br>
y = {{=y}}
<br>
x + y = {{=x+y}}
<hr>
{{lista = range(10)}}
{{for item in lista:}}
	{{if item % 2:}}
		{{=item}} es impar
		<br>
	{{else:}}
		{{=item}} es par
		<br>
	{{pass}}
{{pass}}
</body>
</html>
```

- controllers/default/saluda?nombre=David -> saluda
```python
def saluda():
    'app0/default/saluda?nombre=David'

    # esta acción puede recibir la variable nombre
    nombre = request.vars.get('nombre')

    return locals()
```
- controllers/default/registra -> formulario procesado
```python
def registra():
    'app0/default/registra -> registra y saluda'

    # construyo un formulario con Python + Web2py
    form = FORM(LABEL('Nombre'),
                INPUT(_name='nombre', requires=IS_NOT_EMPTY()),
                INPUT(_type='submit', _value='Registra'))

    # si miras el código fuente HTML verás que el formulario
    # se envía a la misma acción que lo emitió
    # de este modo es la misma acción la que lo procesa
    form.process()  # aquí se procesa el formulario
    if form.accepted:
        # el formulario ha sido aceptado
        # puedo recuperar las variables que he recibido
        nombre = form.vars.nombre
        session.flash = 'Bienvenido ' + nombre
        # redirijo a la acción saluda
        redirect(URL('saluda', vars=dict(nombre=nombre)))
    elif form.errors:
        # hay errores tras procesar el formulario
        response.flash = 'Hay errores en el formulario'
    else:
        # por aquí pasará cuando se cree el formulario
        response.flash = 'Si eres tan amable'

    return locals()
```
- views/default/registra.html
```html
{{extend 'layout.html'}}
<h2>Formulario de registro</h2>
<br>
{{=form}}
<hr>
{{if form.vars:}}
<h2>Variables</h2>
{{=BEAUTIFY(form.vars)}}
{{pass}}
{{if form.errors:}}
<h2>Errores</h2>
{{=BEAUTIFY(form.errors)}}
{{pass}}
```
- controllers/default/registra2 -> formulario desde el modelo procesado
```python
def registra2():
    'app0/default/registra2 -> registra y saluda'

    # construyo un formulario con Python + Web2py
    # pero ahora me apoyo en un modelo simulado
    # Field() es el constructor de campos para los modelos en Web2py
    form = SQLFORM.factory(Field('nombre', requires=IS_NOT_EMPTY()),
                           Field('direccion'),
                           Field('email', requires=IS_EMAIL()),
                           Field('password', 'password', requires=IS_NOT_EMPTY()))

    # si miras el código fuente HTML verás que el formulario
    # se envía a la misma acción que lo emitió
    # de este modo es la misma acción la que lo procesa
    form.process()  # aquí se procesa el formulario
    if form.accepted:
        # el formulario ha sido aceptado
        # redirijo a saluda
        nombre = form.vars.nombre
        session.flash = 'Bienvenido ' + nombre
        redirect(URL('saluda', vars=dict(nombre=nombre)))
    elif form.errors:
        # hay errores tras procesar el formulario
        response.flash = 'Hay errores en el formulario'
    else:
        # por aquí pasará cuando se cree el formulario
        pass

    return locals()
```

## Otra aplicación (app1)
Otra aplicación un poco más compleja sería:

- Modelo app1/models/db.py
```python
db = DAL('sqlite://storage.db')     # conexión con la base de datos
db.define_table('image',            # defino el modelo de datos "image"
    Field('name', notnull=True),    # - campo name
    Field('file','upload'))         # - campo file
```
- Controlador en app1/controllers/default.py
```python
def index():     # http://localhost:8000/app/default/index
    form = SQLFORM(db.image)               # se pueden crear formularios partiendo del modelo
    form.process()                         # procesamiento del formulario
    if form.accepted:                      # cuando el usuario envía un formulario correcto
        response.flash = 'image uploaded'  # se devuelve un mensaje al usuario
    return dict(form=form)                 # la acción devuelve un diccionario => vista
```
- Vista en app1/views/default/index.html
```html
{{extend 'layout.html'}}
<h1>Image upload form</h1>
{{=form}}
```
